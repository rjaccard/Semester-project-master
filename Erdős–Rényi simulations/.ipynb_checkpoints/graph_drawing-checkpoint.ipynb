```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
n = 4
p = 0.6
#random_state = np.random.RandomState(2)
G = nx.erdos_renyi_graph(n, p)
G.nodes[3]["label"] = "Stop"
#G = nx.stochastic_block_model([10,10], [[0.7, 0.1], [0.1, 0.7]], seed=random_state)
nx.draw(G, node_color='r', width=2.0, edge_color='b')
plt.savefig('graph.png', transparent=True)
```

```python
# Create an empty graph
G = nx.Graph()

# Add nodes
G.add_node(1, color='orange')
G.add_node(2, color='orange')
G.add_node(3, color='red')
G.add_node(4, color='red')

# Add edges
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 4)

# Get node colors
node_colors = [G.nodes[node]['color'] for node in G.nodes]

# Draw the network
nx.draw(G, with_labels=True, node_color=node_colors, cmap=plt.get_cmap('coolwarm'), node_size=500, font_size=15, edge_color='w')

# Show the plot
plt.savefig('graph.png', transparent=True)
plt.show()
```
